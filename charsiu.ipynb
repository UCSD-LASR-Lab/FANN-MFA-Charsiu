```python
def convert_lab(lab_file_path, output_txt_file_path):
    try:
        with open(lab_file_path, 'r') as lab_file:
            lab_content = lab_file.read()
    except FileNotFoundError:
        print(f"The file {lab_file_path} does not exist.")
        return

    with open(output_txt_file_path, 'w') as txt_file:
        txt_file.write(lab_content)

    print(f"Conversion completed. Content from {lab_file_path} has been saved to {output_txt_file_path}.")
```

```python
import os
import librosa
import soundfile as sf
from Charsiu import charsiu_forced_aligner

def resample_audio(audio_path, target_sr=16000):
    # Load the audio file
    audio, sr = librosa.load(audio_path, sr=None)
    if sr != target_sr:
        print(f"Resampling from {sr}Hz to {target_sr}Hz.")
        audio_resampled = librosa.resample(audio, orig_sr=sr, target_sr=target_sr)
        sf.write(audio_path, audio_resampled, target_sr)
    else:
        print("Audio is already at the target sampling rate.")

def read_text_from_file(text_path):
    with open(text_path, 'r', encoding='utf-8') as file:
        text = file.read().strip()
    return text

def process_directory(root_dir, output_dir):
    # Initialize Charsiu aligner
    charsiu = charsiu_forced_aligner(aligner='charsiu/en_w2v2_fc_10ms')
    
    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith('.wav'):
                audio_path = os.path.join(dirpath, filename)
                text_path = os.path.splitext(audio_path)[0] + '.txt'
                
                if not os.path.exists(text_path):
                    print(f"Text file for {filename} not found.")
                    continue
                
                # Read the corresponding text for the audio file
                text = read_text_from_file(text_path)
                
                # Resample the audio file
                resample_audio(audio_path)
                
                try:
                    # Perform forced alignment and save the output as a TextGrid file
                    output_path = os.path.join(output_dir, os.path.splitext(filename)[0] + '.TextGrid')
                    charsiu.serve(audio=audio_path, text=text, save_to=output_path)
                    print(f"Processed and saved alignment for {filename} to {output_path}")
                except Exception as e:
                    print(f"\033[91mError processing {filename}: {e}\033[0m")

# Specify the directory containing the .wav and .txt files
root_dir = '/Users/cogsci-lasrlab1/Downloads/MFA_data/KidTalk/pe_ppts'
# Specify the output directory for the TextGrid files
output_dir = '/Users/cogsci-lasrlab1/Downloads/Charsiu-aligned-result/KidTalk/pe_ppts'
os.makedirs(output_dir, exist_ok=True)  # Ensure the output directory exists

process_directory(root_dir, output_dir)
```
