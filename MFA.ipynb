```python
import os

def process_directory(root_dir):
    # Walk through the directory
    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            # Check if the file is a .wav file
            if filename.endswith('.wav'):
                # Extract base name and the word after "_"
                base_name = os.path.splitext(filename)[0]
                try:
                    word_to_write = base_name.split('_')[1]  # Assumes there is at least one "_" in the name
                except IndexError:
                    print(f"No '_' found in {filename}, skipping.")
                    continue
                
                # Create a corresponding .lab file
                lab_file_path = os.path.join(dirpath, base_name + '.lab')
                
                # Write the extracted word into the .lab file
                with open(lab_file_path, 'w') as lab_file:
                    lab_file.write(word_to_write)
                print(f"Created {lab_file_path} with word: {word_to_write}")

# Replace 'your_directory_path_here' with the path to the directory you want to process
process_directory('/Users/cogsci-lasrlab1/Downloads/MFA_data/KidTalk/pe_ppts')
```
